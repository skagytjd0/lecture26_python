### 최대공약수 최소공배수 약수 배수

```python
N = int(input())
cdivisor_list = []
for num in range (1, N + 1): 
    if N % num == 0:
        cdivisor_list.append (num)
print(cdivisor_list)
```

```python
# 최대 공약수

def getGCD(M,N):
    if M > N: M,N = N,M

    for num in range(M,0,-1):
        if (M % num == 0) and (N % num == 0):
            return num
        
print(getGCD(6,12))
```

```python
# 최소 공배수

def getLCM(M,N):
    if M > N: M,N = N,M

    i = 1
    while True:
        multiple = N * i
        if multiple % M == 0:
            return multiple
        i += 1
        
print(getLCM(7,18))
```

### 소수구하기

```python
N = int(input())

print(f"{N} =>" ,end=" ")

is_prime_number = True
if N != 1:
    for i in range(2,N):
        if N % i == 0:
            is_prime_number
            break
else:
    is_prime_number

print("소수" if is_prime_number else "소수가 아님")


```

```python
#진법전환
N = int(input())
base = int(input())

digits = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
result = ""

if N == 0:
    result = "0"
else:
    while N > 0:
        remainder = N % base
        N //= base
        result = digits[remainder] + result

print(result)
```

### 투포인터

```python
#BOJ 11728번(

N,M = map(int,input().split())
A = list(map(int,input().split()))
B = list(map(int,input().split()))
new_list = []

pA = 0
pB = 0
while pA < N and pB < M:
    if A[pA] < B[pB]:
        new_list.append(A[pA])
        pA += 1
    else:
        new_list.append(B[pB])
        pB += 1

while pA < N:
    new_list.append(A[pA])
    pA += 1

while pB < M:
    new_list.append(B[pB])
    pB += 1

for i in new_list:
    print(i, end=" ")
```

```python
# 2003번

N,M = map(int,input().split())
num_list = list(map(int,input().split()))
count = 0
start = end = 0
sum = num_list[0]
while end < N:
    if sum == M: count += 1
    if sum <= M:
        end += 1
        if end < N: 
            sum += num_list[end]
    else:
        sum -= num_list[start]
        start += 1
print(count)
```
