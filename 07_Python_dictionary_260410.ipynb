### 딕셔너리 실습

```python
coffee_menu = {'아메리카노': 2000, '라떼': 3000, '녹차': 2500}



if '카푸치노' not in coffee_menu:
    coffee_menu['카푸치노'] = 3500

names = list(coffee_menu.keys())
prices = list(coffee_menu.values())
items = list(coffee_menu.items())

for name in coffee_menu:
    val = coffee_menu.get(name)
    print(f"메뉴명: {name: <6} | 가격: {val:,}원")

removed_val = coffee_menu.pop('녹차')
del coffee_menu['아메리카노']

print(names)
print(prices)
print(coffee_menu)

coffee_menu.clear()
print(f"초기화 완료: {coffee_menu}")
```
